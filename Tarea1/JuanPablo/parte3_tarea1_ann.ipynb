```python
import numpy as np
import keras
from sklearn.preprocessing import StandardScaler
from sklearn import cross_validation, model_selection
from keras import backend as K
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.layers import Dropout
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta
from keras import regularizers
import scipy.io as sio

import matplotlib.pyplot as plt
%matplotlib inline

mat_test = sio.loadmat("AgeGenderClassification/eventest.mat")
mat_train = sio.loadmat("AgeGenderClassification/eventrain.mat")
data_test = mat_test["tecoll"][0][0]
data_train = mat_train["trcoll"][0][0]
```

```python
for i in range(5):
    print(data_train[i].shape)
```

```python
scaler = StandardScaler().fit(data_train[0])
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train = scaler.transform(data_train[0])
Y_train = data_train[1]
X_test = scaler.transform(data_test[0])
Y_test = data_test[1]
y_train=Y_train
y_test=Y_test

```

```python
modelf = Sequential()
modelf.add(Dense(10, input_dim=14, kernel_initializer='uniform',activation='sigmoid'))
modelf.add(Dense(8, kernel_initializer='uniform',activation='sigmoid'))
modelf.add(Dense(6, kernel_initializer='uniform',activation='sigmoid'))
modelf.add(Dense(4, kernel_initializer='uniform',activation='sigmoid'))
modelf.add(Dense(1, kernel_initializer='uniform',activation='linear'))
sgd = Adam(lr=0.01)
modelf.compile(optimizer=sgd,loss='mean_squared_error')
```

```python
modelf.fit(X_train,y_train,epochs=100,batch_size=50,verbose=0,shuffle=True)
```

```python
score=modelf.evaluate(X_test,y_test,verbose=1)
print("\n Loss: %.3f"%(score))
```

```python
#Uso de kfold para encontrar los mejores hiperparametros a utilizar
optimizadores =[Adagrad(lr=0.01), Adadelta(lr=0.01), RMSprop(lr=0.01), Adam(lr=0.01)]
Xm = X_train
ym = y_train
kfold = cross_validation.KFold(len(Xm), 5)
cvscores = []
all_mse_cv = []
lr_rates = [0.01, 0.1, 0.3, 0.5, 0.7]
for opt in optimizadores:
    for i, (train, val) in enumerate(kfold):
        model = Sequential()
        model.add(Dense(10, input_dim=X_train.shape[1], kernel_initializer='uniform',activation="sigmoid"))
        model.add(Dense(8, kernel_initializer='uniform',activation="sigmoid"))
        model.add(Dense(6, kernel_initializer='uniform',activation="sigmoid"))
        model.add(Dense(4, kernel_initializer='uniform',activation="sigmoid"))
        model.add(Dense(1, kernel_initializer='uniform',activation="linear")) 
        model.compile(optimizer=opt,loss='mean_squared_error')
        model.fit(Xm[train], ym[train], epochs=50, verbose=1)
        scores = model.evaluate(Xm[val], ym[val])
        cvscores.append(scores)
    mse_cv = np.mean(cvscores)
    all_mse_cv.append(mse_cv)
indexModel = all_mse_cv.index(min(all_mse_cv))
print("Mejor optimizador: ", indexModel )
```

```python
Xm = X_train
ym = y_train
kfold = cross_validation.KFold(len(Xm), 5)
cvscores = []
all_mse_cv = []
activaciones = [['sigmoid','sigmoid','sigmoid'],['sigmoid','sigmoid','relu'],['sigmoid','relu','sigmoid'],['sigmoid','relu','relu'],['relu','sigmoid','sigmoid'],['relu','sigmoid','relu'],['relu','relu','sigmoid'],['relu','relu','relu']]
for act in activaciones:
    for i, (train, val) in enumerate(kfold):
        model = Sequential()
        model.add(Dense(10, input_dim=X_train.shape[1], kernel_initializer='uniform',activation=act[0]))
        model.add(Dense(8, kernel_initializer='uniform',activation=act[1]))
        model.add(Dense(6, kernel_initializer='uniform',activation=act[2]))
        model.add(Dense(4, kernel_initializer='uniform',activation="sigmoid"))
        model.add(Dense(1, kernel_initializer='uniform',activation="linear")) 
        model.compile(optimizer=optimizadores[indexModel],loss='mean_squared_error')
        model.fit(Xm[train], ym[train], epochs=50, verbose=1)
        scores = model.evaluate(Xm[val], ym[val])
        cvscores.append(scores)
    mse_cv = np.mean(cvscores)
    all_mse_cv.append(mse_cv)
indexModel = all_mse_cv.index(min(all_mse_cv))
print("Mejor optimizador: ", indexModel )
```
